```python
'''
As hammoq provides api to fetch historical data posted by customers.
Analysing that data to look for most frequent attributes and their corresponding values 
can help us to decide on keywords that we chose to scrap similar data from other platforms

This notebook provides methods to process json files coming from hammoq api, build dataframe out of these files and
perform histogram based visual analysis for attributes and corresponding values.

'''
```

```python
import os
import pandas as pd
from collections import Counter
import json
from matplotlib import pyplot as plt
```

```python
def get_dataframe_from_json_folder(json_folder_path='./'):
    '''
    read json files from given path and build list of dictionaries base on json strings
    '''
    json_folder_path = json_folder_path
    all_jsons=[]
    for root, directories, files in os.walk(json_folder_path, topdown=False):
        for name in files:
            f=(os.path.join(root, name))
            if f.endswith((".json")):
                all_jsons.append(f)
    '''
    read json files and build list of dictionaries base on json strings
    '''

    dcts=[]
    for json_file in all_jsons:
        with open(json_file,'r') as f:
            data= eval(json.load(f))
            dcts.append(data)
    '''
    build dataframe from list of dictionaries
    '''
    _df=pd.DataFrame(dcts)
    return _df
                                                                
    
```

### methods for attributes and corresponding values analysis


```python
'''
method to get list of attributes with their frequencies in given columns of dataframe
input: dataframe, name of column {default=RECOMMENDED}
output: attributes frequency tabel
'''
def get_attributes_from_df(dataframe,column_name='RECOMMENDED'):
    lsts = []
    dcts=[]
    for datums in dataframe[column_name]:
        lsts.append(datums)
        for dct in datums:
            dcts.append(dct)
    attributes=[]
    for lst in lsts:
        for dct in lst:
            for key in dct.keys():
                attributes.append(key)
    #attrib_freq=sorted(Counter(attributes).items(), key=lambda item: item[1],reverse=True)
    return list(Counter(attributes).keys())

def get_attributes_freq_from_df(dataframe,column_name='RECOMMENDED'):
    lsts = []
    dcts=[]
    for datums in dataframe[column_name]:
        lsts.append(datums)
        for dct in datums:
            dcts.append(dct)
    attributes=[]
    for lst in lsts:
        for dct in lst:
            for key in dct.keys():
                attributes.append(key)
    return sorted(Counter(attributes).items(), key=lambda item: item[1],reverse=True)
    
```

```python
def get_vals_for_attrib(attrib,dataframe,column_name='RECOMMENDED'):
    dcts = []
    vals = []
    for lst in dataframe[column_name]:
        for dct in lst:
            dcts.append(dct)
    for i in dcts:
        if attrib in list(i.keys()):
            if isinstance(i[attrib],str):
                vals.append(i[attrib])
            elif isinstance(i[attrib],list):
                for val in i[attrib]:
                    vals.append(val)
            else:
                pass
    return list(Counter(vals).keys())
```

## driver method to write attributes and corresponding values to csv file

```python
def gen_attributes_csv_from_df(df,column_name,filename=None,thresh_hold=500):
    attribs=get_attributes_from_df(df,column_name)
    #attribs_dct = {}
    attribs_vals = []
    for attrib in attribs:
        vals = get_vals_for_attrib(attrib,df,column_name)
        attribs_vals.append(vals)
        #attribs_dct[attrib]=vals
    # generate datframe for lst
    attribs_vals_df=pd.DataFrame(attribs_vals,index=attribs).T
    if filename is not None:
        attribs_vals_df.to_csv(filename)
    return attribs_vals_df
```
